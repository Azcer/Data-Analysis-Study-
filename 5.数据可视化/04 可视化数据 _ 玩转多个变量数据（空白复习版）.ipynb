```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 给散点图增加变量

```python
# 导入本节示例数据，注意此文件并非本地数据文件，而是从Seaborn库里导入的
# Seaborn库里的所有数据可以在这里查看：https://github.com/mwaskom/seaborn-data
# 如果遇到导入问题，可以从以上GitHub仓库下载到本地，然后读取文件
# 或者是参考https://blog.csdn.net/weixin_51755670/article/details/127979895
iris = pd.read_csv("iris.csv")
iris
```

```python
# 传入iris作为数据，以sepal_length为横轴数据，sepal_width为纵轴数据，绘制散点图
sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'])
plt.show()
```

```python
# 传入iris作为数据，以sepal_length为横轴数据，sepal_width为纵轴数据，点的颜色为绿色，绘制散点图
sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'], color='green')
plt.show()
```

```python
# 传入iris作为数据，以sepal_length为横轴数据，sepal_width为纵轴数据，点的颜色根据species分类，绘制散点图
sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'], hue=iris['species'])
plt.show()
```

```python
# 传入iris作为数据，以sepal_length为横轴数据，sepal_width为纵轴数据，点的颜色根据petal_length数值，绘制散点图
sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'], hue=iris['petal_length'])
plt.show()
```

```python
# 传入iris作为数据，以sepal_length为横轴数据，sepal_width为纵轴数据，点的颜色根据species分类
# 点的尺寸根据petal_length数值，绘制散点图
sns.scatterplot(x=iris['sepal_length'], y=iris['sepal_width'], hue=iris['species'], size=iris['petal_length'])
plt.show()
```

```python
sns.scatterplot(iris, x="sepal_length", y="sepal_width", 
                hue="species", size="petal_length")
# 添加一行代码，把图例放在左上方
plt.legend(bbox_to_anchor=(0,1))
plt.show()
```

```python
sns.scatterplot(iris, x="sepal_length", y="sepal_width", 
                hue="species", size="petal_length")
# 添加一行代码，把图例放在右下方
plt.legend(bbox_to_anchor=(1,0))
plt.show()
```

```python
sns.scatterplot(iris, x="sepal_length", y="sepal_width", 
                hue="species", size="petal_length")
# 添加一行代码，把图例放在右上方
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

## 给折线图增加变量

```python
# 导入示例数据
flights = pd.read_csv("flights.csv")
flights
```

```python
# 传入flights作为数据，以month为横轴数据，passengers为纵轴数据，点的颜色根据year数值，绘制折线图
# 图例放在右上方
sns.lineplot(x=flights['month'], y=flights['passengers'], hue=flights['year'])
plt.legend(bbox_to_anchor=(1,1))
```

## 给条形图增加变量

```python
# 导入示例数据
penguins = pd.read_csv("penguins.csv")
penguins
```

```python
# 传入penguins作为数据，以species为横轴数据，body_mass_g为纵轴数据，条柱的颜色根据sex分类，绘制复合直方图
sns.barplot(x=penguins['species'], y=penguins['body_mass_g'], hue=penguins['sex'])
plt.show()
```

## 热力图

```python
# 导入示例数据
glue = pd.read_csv("glue.csv").pivot(index='Model', columns='Task', values='Score')
glue
```

```python
# 传入glue作为数据，绘制热力图

sns.heatmap(glue, annot=True)
plt.show()
```

## 图上绘制多个图表

```python
# 对所需数据进行赋值
setosa = iris.query('species == "setosa"')
versicolor = iris.query('species == "versicolor"')
virginica = iris.query('species == "virginica"')
```

```python
# 在同一张图上对setosa、versicolor、virginica的petal_length变量绘制直方图
sns.histplot(setosa['petal_length'])
sns.histplot(versicolor['petal_length'])
sns.histplot(virginica['petal_length'])
plt.show()
```

```python
# 在同一张图上对setosa、versicolor、virginica的petal_length变量绘制直方图，条柱宽度为0.1

sns.histplot(setosa['petal_length'], binwidth=0.1)
sns.histplot(versicolor['petal_length'], binwidth=0.1)
sns.histplot(virginica['petal_length'], binwidth=0.1)
plt.show()
```

```python
# 在同一张图上对setosa、versicolor、virginica的petal_length变量绘制直方图，条柱宽度为0.1
# 在图上给三个直方图添加标签，分别为“Setosa”、“Versicolor”、“Virginica”，并展示图例

sns.histplot(setosa['petal_length'], binwidth=0.1, label='setosa')
sns.histplot(versicolor['petal_length'], binwidth=0.1, label='versicolor')
sns.histplot(virginica['petal_length'], binwidth=0.1, label='virginica')
plt.legend()
plt.show()
```

```python
# 在同一张图上对setosa、versicolor、virginica的petal_length变量绘制密度图，条柱宽度为0.1
# 在图上给三个密度图添加标签，分别为“Setosa”、“Versicolor”、“Virginica”，并展示图例
sns.kdeplot(setosa['petal_length'], label='setosa')
sns.kdeplot(versicolor['petal_length'], label='versicolor')
sns.kdeplot(virginica['petal_length'],label='virginica')
plt.legend()
plt.show()
```

```python
# 在同一张图上对setosa、versicolor、virginica的petal_length变量绘制箱型图
sns.boxplot(setosa['petal_length'])
sns.boxplot(versicolor['petal_length'])
sns.boxplot(virginica['petal_length'])
plt.show()
```

```python
# 在同一张图上对setosa、versicolor、virginica的petal_length变量绘制小提琴图

sns.violinplot(setosa['petal_length'])
sns.violinplot(versicolor['petal_length'])
sns.violinplot(virginica['petal_length'])
plt.show()
```

```python
# 传入iris作为数据，以species为横轴数据，petal_length为纵轴数据，绘制箱型图
sns.boxplot(y=iris['petal_length'], x=iris['species'])
plt.show()
```

```python
# 传入iris作为数据，以species为横轴数据，petal_length为纵轴数据，绘制小提琴图
sns.violinplot(y=iris['petal_length'], x=iris['species'])
plt.show()
```

## 图上绘制多个子图

```python
# 绘制一个一行三列的空白子图
plt.subplots(1,3)
plt.show()
```

```python
# 绘制一个一行三列的空白子图，子图尺寸为(15 ,5)
plt.subplots(1,3, figsize=(15,5))
plt.show()
```

```python
# 绘制一个一行三列的子图，子图尺寸为(15 ,5)
# 每个子图里传入iris作为数据，species为横轴变量，纵轴变量在每个子图里分别为sepal_length、sepal_width、petal_length，绘制箱型图
fig, axes = plt.subplots(1,3, figsize=(15,5))
sns.boxplot(x=iris['species'], y=iris['sepal_length'], ax=axes[0])
sns.boxplot(x=iris['species'], y=iris['sepal_width'], ax=axes[1])
sns.boxplot(x=iris['species'], y=iris['petal_length'], ax=axes[2])
plt.show()
```

## 成对图（Pair Plot）

```python
# 传入iris作为数据，绘制成对图（Pair Plot）
sns.pairplot(iris)
plt.show()
```

```python
# 传入iris作为数据，以species进行颜色分类，绘制成对图（Pair Plot）

sns.pairplot(iris, hue='species')
plt.show()
```
